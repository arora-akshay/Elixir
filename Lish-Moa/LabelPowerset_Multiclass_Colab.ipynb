```python
from skmultilearn.problem_transform import LabelPowerset
from sklearn.naive_bayes import GaussianNB

import pandas as pd

from sklearn.metrics import accuracy_score
```

```python
trainds = pd.read_csv('/content/train_features.csv',index_col='sig_id')

labelds = pd.read_csv('/content/train_targets_scored.csv',index_col='sig_id')

testds = pd.read_csv('/content/test_features.csv',index_col='sig_id')
```

```python
trainds.head()
```

```python
labelds.head()
```

```python
trainds['cp_type'] = trainds['cp_type'].map({'trt_cp':0,'ctl_vehicle':1})

trainds['cp_dose'] = trainds['cp_dose'].map({'D1':0,'D2':1})
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(trainds, labelds, test_size=0.10, random_state=42)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis

h = .02  # step size in the mesh

names = ["Nearest Neighbors", "Linear SVM", "RBF SVM", 
         "Gaussian Process",
         "Decision Tree", "Random Forest", "Neural Net", "AdaBoost",
         "Naive Bayes", "QDA"]

classifiers = [
    KNeighborsClassifier(),
    SVC(kernel="linear", C=0.025),
    SVC(gamma='auto', C=1),
    GaussianProcessClassifier(1.0 * RBF(1.0)),
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    MLPClassifier(),
    AdaBoostClassifier(),
    GaussianNB()]

```

```python
output = {}
```

```python
for name, clf in zip(names, classifiers):
  classifier = LabelPowerset(clf)
  classifier.fit(X_train, y_train)
  predictions = classifier.predict(X_test)
  accuracyscore = accuracy_score(y_test,predictions)
  output.update({name:accuracyscore})
  print(output)
```
