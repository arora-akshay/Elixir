```python
from skmultilearn.problem_transform import LabelPowerset
from sklearn.naive_bayes import GaussianNB

import pandas as pd

from sklearn.metrics import accuracy_score
```

```python
trainds = pd.read_csv('/content/train_features.csv',index_col='sig_id')

labelds = pd.read_csv('/content/train_targets_scored.csv',index_col='sig_id')

testds = pd.read_csv('/content/test_features.csv',index_col='sig_id')
```

```python
trainds.head()
```

```python
labelds.head()
```

```python
trainds['cp_type'] = trainds['cp_type'].map({'trt_cp':0,'ctl_vehicle':1})

trainds['cp_dose'] = trainds['cp_dose'].map({'D1':0,'D2':1})
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(trainds, labelds, test_size=0.10, random_state=42)
```

```python
classifier = LabelPowerset(GaussianNB())

classifier.fit(X_train, y_train)
```

```python
predictions = classifier.predict(X_test)

accuracy_score(y_test,predictions)
```

```python
from sklearn.ensemble import RandomForestClassifier

classifier = LabelPowerset(RandomForestClassifier())

classifier.fit(X_train, y_train)
```

```python
predictions = classifier.predict(X_test)

accuracy_score(y_test,predictions)
```
